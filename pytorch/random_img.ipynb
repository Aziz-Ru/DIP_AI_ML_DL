```python
serial = torch.arange(start=0,end=12,step=3)
# serial*10
# serial+10
# serial/2
addition =torch.add(serial,100)
# print(addition)
subtraction = torch.sub(addition,10)
# print(subtraction)
multiplication = torch.mul(subtraction,10)
# print(multiplication)
division = torch.div(multiplication,100)
# print(division)

log = torch.log(division)
# print(log)
exponential = torch.exp(log)
# print(exponential)

# print(torch.max(exponential))
# print(torch.min(exponential))
# print(torch.argmax(exponential))
# print(torch.argmin(exponential))
# print(torch.mean(exponential))
# print(torch.std(exponential))
# print(torch.var(exponential))
# print(torch.sum(exponential))
# print(torch.unique(exponential))

tensor = torch.arange(0,100,10,dtype=torch.float32)
print(tensor)
print(f'Max : {tensor.max()}')
print(f'Min : {tensor.min()}')
print(f'Argmax : {tensor.argmax()}')
# Returns the index of the maximum value.
print(f'Argmin : {tensor.argmin()}')
# Returns the index of the minimum value.
print(f'Mean : {tensor.mean()}')
print(f'Std : {tensor.std()}')
print(f'Var : {tensor.var()}')
print(f'Sum : {tensor.sum()}')
print(f'Unique : {tensor.unique()}')
```
