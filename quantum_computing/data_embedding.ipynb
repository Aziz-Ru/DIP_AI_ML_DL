```python
import pennylane as qml
from pennylane import numpy as np
```

Basis Encoding
Binary to qunatum state

```python
wires = range(2)
# range is a Python function that generates a sequence of integers from 0 up to (but not including) 2. So, range(2) produces [0, 1]
dev = qml.device("default.qubit", wires=wires)
# In PennyLane, "wires" represent the qubits in a quantum device

#  data type must be integers
# encode n binary featyre into basis state of n qubits
x = np.array([1,0], dtype=int)
@qml.qnode(dev)
def basis_encoding(x):
    qml.BasisEmbedding(features=x,wires=wires)
    return qml.state()

basis_encoding(x)
```
